```python
df['target']=df.shift(-1)['temp_max (⁰C)']
df
```

```python
from sklearn.linear_model import Ridge
rr = Ridge(alpha=.1)
```

```python
predictors = df.columns[~df.columns.isin(['target'])]
#predictors = df.columns
```

```python
def backtest(df, model, predictors, start=750, step=30):
    all_predictions = []
    
    for i in range(start, df.shape[0], step):
        train = df.iloc[:i,:]
        test = df.iloc[i:(i+step),:]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions_rr = backtest(df, rr, predictors)
```

```python
predictions_rr
```

```python
predictions1_rr=predictions_rr.copy()
```

```python
predictions1_rr.drop(['diff'],axis=1,inplace=True)
```

```python
predictions1_rr
```

```python
predictions1_rr.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae_rr=mean_absolute_error(predictions_rr["prediction"], predictions_rr["prediction"])
mae_rr
```
